```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa=pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
#to get column name of each column
for col in fifa.columns:
    print(col)
```

```python
#number of rows and col
fifa.shape
```

```python
#18278 players
```

```python
#to find how many players belong to each nationality
fifa['nationality'].value_counts()
```

```python
#to find top 10 nationality having max players
fifa['nationality'].value_counts()[0:10]
```

```python
#to find name of top 5 nationality having max players
list(fifa['nationality'].value_counts().keys()[0:10])
```

```python
#bar graph visualization of countries having max players
plt.figure(figsize=(7,4))
plt.bar(list(fifa['nationality'].value_counts().keys()[0:5]),list(fifa['nationality'].value_counts()[0:5]),color=['red','green','yellow','blue','purple'])
plt.title("Nationality v/s Num of Players")
plt.xlabel("Nationality")
plt.ylabel("Number of Players")
plt.show()
```

```python
#pie chart visualization of countries having max players
plt.pie(list(fifa['nationality'].value_counts()[0:5]),labels=list(fifa['nationality'].value_counts().keys()[0:5]),autopct='%0.1f%%')
plt.title("Pie chart for number of players per nationality")
plt.show()
```

```python
#to extract only name and salary of players
salary=fifa[['short_name','wage_eur']]
```

```python
salary.head()
```

```python
#to sort by salary
salary=salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
salary.head()
```

```python
#to get bar graph of player vs salary
plt.figure(figsize=(9,4))
plt.bar(list(salary['short_name'][0:5]),list(salary['wage_eur'][0:5]))
plt.title("Name v/s Salary")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.show()
```
